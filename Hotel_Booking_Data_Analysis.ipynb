# Detect outliers using IQR

```python
# Function to detect outliers using IQR
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

# Loop through all numerical columns and check for outliers
for col in numerical_cols:
    outliers = detect_outliers_iqr(df, col)
    print(f"{col}: {outliers.shape[0]} outliers found")
```

```python
# Calculates the average daily rate (ADR) for canceled bookings, grouped by reservation date.
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()

# Resets index to make 'reservation_status_date' a column.
cancelled_df_adr.reset_index(inplace = True)

# Ensures data is sorted by reservation date for proper trend visualization.
cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

# Filters data for reservations that were not canceled.
not_cancelled_data = df[df['is_canceled'] == 0]

# Calculates ADR for non-canceled bookings.
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()

# Resets index and sorts data for proper visualization.
not_cancelled_df_adr.reset_index(inplace = True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

# Plots ADR trends for both canceled and non-canceled bookings over time.
plt.figure(figsize = (20,6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'], label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'], label = 'cancelled')
plt.legend()
```

* There are spikes in ADR, especially for canceled reservations, which indicates that higher-priced bookings are more likely to be canceled.
*This suggests that customers who book at higher prices tend to cancel more often, possibly due to affordability concerns or alternative options becoming available.
* Over time, both canceled and not canceled ADR trends increase, possibly due to seasonal effects or price changes.

*Suggestions:
* Stricter Cancellation Policies
*Incentives for Non-Cancellations

# # Filtering ADR Data for a Specific Time Period (2016 to August 2017)

```python
# Filtering the cancelled bookings ADR data within the specified time range
cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016') & (cancelled_df_adr['reservation_status_date']<'2017-09')]

# Filtering the non-cancelled bookings ADR data within the specified time range
not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016') & (not_cancelled_df_adr['reservation_status_date']<'2017-09')]
```

```python
plt.figure(figsize = (20,6))
plt.title('Average Daily Rate', fontsize = 30)

# Plotting ADR for non-cancelled bookings
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'], label = 'not cancelled')

# Plotting ADR for cancelled bookings
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'], label = 'cancelled')
plt.legend(fontsize = 20)
plt.show()
```

# insights
* Cancelled bookings generally had a higher ADR than non-cancelled ones throughout 2016 and 2017, suggesting price sensitivity.
*Both categories showed an upward trend in ADR from early 2016 to mid-2017, likely due to seasonal demand and pricing adjustments.
